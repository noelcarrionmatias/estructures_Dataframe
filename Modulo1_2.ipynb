```python
#Funciona! Velocidad de los aviones
vuelos = data3.iloc[:,[10,12,19]]
tiempo = vuelos['ActualElapsedTime']/60
distancia = vuelos['Distance']
kmh = distancia/tiempo

#No consigo agregar el nombre del vuelo y juntar vuelo-velocidad
#vuelos['velocidad'] = kmh
#vuelos.loc[0,3]= kmh
#print(vuelos)

#Preguntamos qué quiere ver y el input()
print('Velocidad de aviones:\n Clica el siguiente número para ver:\n 1 - Velocidad máximan\n 2 - Velocidad media\n 3 - Velocidad mínima')
x = int(input())

#Definimos el resultado final
if x == 1:
    print('La velocidad máxima es de:',kmh.max(),'km/h')
elif x == 2:
    print('La velocidad media es de :',kmh.mean(),'km/h')
else:
    print('La velocidad mínima es de:',kmh.min(),'km/h')

```

```python
#Funciona! Visualizar los aviones que salen con retraso o antes (interacción)
#preguntamos que quieren hacer:
print('Selecciona 1-2-3 para ver cuantos...: \n1 - Aviones con salida antes de hora, 2 - Aviones con salida despues de hora, 3 - Aviones con salida a la hora\n')
x = int(input())

#definimos las categorias y que comienzen a contar desde 0:
contno=0
contsi=0
contigual=0

#función para calcular los valores mayores, menores o igual
for i in range(0,1936758):
    if data3.iloc[i,[7][0]] < data3.iloc[i,[8][0]]:
        contno=contno+1
    elif data3.iloc[i,[7][0]] > data3.iloc[i,[8][0]]:
        contsi=contsi+1
    else:
        contigual=contigual+1

#representación de los resultados segun la decisión del imput():
if x == 1:
        print('Hay:',contno,'aviones que han salido antes de hora')
elif x == 2:
        print('Hay:',contsi,'aviones que han salido despues de hora')
else:
        print('Hay:',contigual,'aviones que han salido puntuales')
```

```python
#Funciona! agrupar vuelos por meses:
mes = data3.iloc[:,[2,10]]
mes1 = mes.groupby(['Month']).count()
mes2 = mes1.sort_values(by=(['FlightNum']),ascending=False)
print(mes2)
```

```python
#Funciona! Ranking en número de horas en retrasos por compañia
retraso = data2.iloc[:,[9,29]]
retraso1 = retraso.groupby('UniqueCarrier').sum().rank(ascending=False)

#añadimos una columna nueva con el total de horas de retraso y ordenamos por horas de retraso
retraso1.loc[:,'horas de retraso']= retraso.groupby('UniqueCarrier').sum()
retraso2 = retraso1.sort_values(by=(['LateAircraftDelay']),ascending=True)

print(retraso2)
```
